```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/MarketPricePrediction.csv')

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Sort the DataFrame by the 'date' column in ascending order
df = df.sort_values(by='date')

# Set the 'date' column as the index
df.set_index('date', inplace=True)
# Check for non-numeric values in 'Modal_price' column
df = df.drop(columns=['market', 'month', 'year','state', 'city'])

# Resample the data to monthly frequency and take the mean
df_resampled = df.resample('M').mean()

# Display the first few rows of the sorted DataFrame
df_resampled.head()
```

```python
# Check for missing values
df_resampled.info()

# Describe the dataset to understand its statistical properties
df_resampled.describe()

# Plotting the time series data for price and quantity
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(df_resampled['priceMod'], label='Mode Price')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Quantity and Mode Price Over Time')
plt.legend()
plt.show()
```

# Forecasting  Price

## ARIMA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
from tqdm import tqdm

# Ignore warnings
warnings.filterwarnings("ignore")

# Sample data preparation (you need to replace this with your actual data)
# Assuming `data` is a DataFrame with a datetime index and a 'quantity' column
# data = pd.read_csv('your_data.csv', index_col='date', parse_dates=True)

# Define the range for p, d, q
p_values = range(0, 15)
d_values = range(0, 5)
q_values = range(0, 11)

# Initialize variables to store the best parameters and the corresponding metric value
best_score, best_cfg = float("inf"), None
warnings.filterwarnings("ignore")

# Grid search
for p in tqdm(p_values):
    for d in d_values:
        for q in q_values:
            try:
                # Define and fit the ARIMA model
                model = ARIMA(df_resampled['priceMod'], order=(p, d, q))
                result = model.fit()

                # Forecasting
                steps = 12  # Forecasting 12 steps ahead
                forecast = result.forecast(steps=steps)

                # Ensure the forecast length matches the actual data length for comparison
                if len(forecast) == len(df_resampled['priceMod'][-steps:]):
                    # Calculate evaluation metrics
                    mae = mean_absolute_error(df_resampled['priceMod'][-steps:], forecast)
                    mse = mean_squared_error(df_resampled['priceMod'][-steps:], forecast)
                    rmse = np.sqrt(mse)

                    # Use AIC as the evaluation metric for model selection
                    aic = result.aic

                    # Print progress
                    print(f'ARIMA({p},{d},{q}) - AIC:{aic}, MAE:{mae}, MSE:{mse}, RMSE:{rmse}')

                    # Select the best model based on AIC
                    if aic < best_score:
                        best_score, best_cfg = aic, (p, d, q)
            except Exception as e:
                continue

print(f'Best ARIMA{best_cfg} - AIC:{best_score}')

# Fit the best ARIMA model
best_p, best_d, best_q = best_cfg
best_model = ARIMA(df_resampled['priceMod'], order=(best_p, best_d, best_q))
best_result = best_model.fit()

# Forecasting with the best model
best_forecast = best_result.forecast(steps=12)

# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(df_resampled['priceMod'], label='Actual priceMod')
plt.plot(best_forecast, label='Best ARIMA Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('priceMod')
plt.title('Best ARIMA Model Forecast')
plt.legend()
plt.show()

# Evaluate the best ARIMA model
best_mae = mean_absolute_error(df_resampled['priceMod'][-12:], best_forecast)
best_mse = mean_squared_error(df_resampled['priceMod'][-12:], best_forecast)
best_rmse = np.sqrt(best_mse)
print(f'Best ARIMA MAE: {best_mae}, MSE: {best_mse}, RMSE: {best_rmse}')

```

## VAR

```python
from statsmodels.tsa.api import VAR

# Select endogenous variables for VAR
var_data = df_resampled[['quantity', 'priceMod']]

# Fit VAR model
var_model = VAR(var_data)
var_result = var_model.fit(maxlags=25, ic='aic')

# Forecasting
var_forecast = var_result.forecast(var_data.values[-var_result.k_ar:], steps=12)

# Convert forecast to DataFrame
var_forecast_df = pd.DataFrame(var_forecast, index=pd.date_range(start=var_data.index[-1], periods=12, freq='M'), columns=var_data.columns)


# Plot forecast
plt.figure(figsize=(10, 6))
plt.plot(df_resampled['priceMod'], label='Actual price')
plt.plot(var_forecast_df['priceMod'], label='VAR Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('priceMod')
plt.title('VAR Model Forecast')
plt.legend()
plt.show()
# Evaluate VAR model
var_mae = mean_absolute_error(df_resampled['priceMod'][-12:], var_forecast_df['priceMod'])
var_mse = mean_squared_error(df_resampled['priceMod'][-12:], var_forecast_df['priceMod'])
var_rmse = np.sqrt(var_mse)
print(f'VAR MAE: {var_mae}, MSE: {var_mse}, RMSE: {var_rmse}')
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(df_resampled[['priceMod','quantity']])

# Prepare training and testing datasets
train_size = int(len(data_scaled) * 0.8)
train, test = data_scaled[:train_size], data_scaled[train_size:]

# Function to create a dataset with look-back
def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset) - look_back - 1):
        a = dataset[i:(i + look_back), :]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

look_back = 1
X_train, Y_train = create_dataset(train, look_back)
X_test, Y_test = create_dataset(test, look_back)

# Reshape input to be [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], X_test.shape[2]))

# Create and fit the LSTM network
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(look_back, 2)))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, Y_train, epochs=10, batch_size=5, verbose=1)

# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Inverse transform predictions
train_predict = scaler.inverse_transform(np.concatenate((train_predict, np.zeros((train_predict.shape[0], 1))), axis=1))[:, 0]
test_predict = scaler.inverse_transform(np.concatenate((test_predict, np.zeros((test_predict.shape[0], 1))), axis=1))[:, 0]

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(df_resampled.index[:train_size], df_resampled['priceMod'][:train_size], label='Train Actual price')
plt.plot(df_resampled.index[train_size:], df_resampled['priceMod'][train_size:], label='Test Actual price')
plt.plot(df_resampled.index[train_size + look_back + 1:], test_predict, label='LSTM Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('price')
plt.title('LSTM Model Forecast')
plt.legend()
plt.show()

# Evaluate LSTM model
lstm_mae = mean_absolute_error(df_resampled['priceMod'][train_size + look_back + 1:], test_predict)
lstm_mse = mean_squared_error(df_resampled['priceMod'][train_size + look_back + 1:], test_predict)
lstm_rmse = np.sqrt(lstm_mse)
print(f'LSTM MAE: {lstm_mae}, MSE: {lstm_mse}, RMSE: {lstm_rmse}')
```

```python
print(f'ARIMA MAE: {arima_mae}, MSE: {arima_mse}, RMSE: {arima_rmse}')
print(f'VAR MAE: {var_mae}, MSE: {var_mse}, RMSE: {var_rmse}')
print(f'LSTM MAE: {lstm_mae}, MSE: {lstm_mse}, RMSE: {lstm_rmse}')
```

```python
# Choose the model with the lowest RMSE
if arima_rmse < var_rmse and arima_rmse < lstm_rmse:
    print("ARIMA is the champion model.")
elif var_rmse < arima_rmse and var_rmse < lstm_rmse:
    print("VAR is the champion model.")
else:
    print("LSTM is the champion model.")
```
